## Comparação entre idade e status de aprovação

```python
plt.figure(figsize=(10, 6))
sns.stripplot(data=df, x='status de aprovacao', y='idade', jitter=True, palette='Set2')

# Adicionando rótulos e título
plt.xticks([0, 1], ['Rejeitado', 'Aprovado'])
plt.xlabel('Status de Aprovação')
plt.ylabel('Idade')
plt.title('Comparação entre Idade e Status de Aprovação')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Exibindo o gráfico
plt.show()
```

```python
df['genero'] = df['genero'].replace({'male': 'masculino', 'female': 'feminino'})
df.head(1)
```

```python
df['nivel educacional'] = df['nivel educacional'].replace({'Master': 'Mestre',
                                                           'High School': 'Ensino Médio',
                                                           'Bachelor': 'Bacharel',
                                                           'Associate': 'Associado',
                                                           'Doctorate': 'Doutorado'
                                                          })
df.head(1)
```

```python
df['situacao casa'] = df['situacao casa'].replace({'RENT': 'Aluguel',
                                                           'OWN': 'Casa própria',
                                                           'MORTGAGE': 'Hipotéca',
                                                           'OTHER': 'Outros',
                                                          })
df.head(1)
```

```python
df['finalidade do emprestimo'] = df['finalidade do emprestimo'].replace({'PERSONAL': 'Pessoal',
                                                   'EDUCATION': 'Educação',
                                                   'MEDICAL': 'Médica',
                                                   'VENTURE': 'Risco',
                                                   'HOMEIMPROVEMENT': 'Melhoria de Casa',
                                                   'DEBTCONSOLIDATION': 'Consolidação de Dívidas',
                                                    })
df.head(1)
```

```python
df['inadimplencia'] = df['inadimplencia'].replace({'No': 'Não', 'Yes': 'Sim'})
df.head(1)
```

```python
df.to_csv('df_tratado.csv', index=False)
```

## Relatório

- Tem pessoas com idade muita avançada fazendo empréstimo em sempre é negado identificando possivel fraude
- Masculino e Ferminino tem mais empréstimo negado do que aprovado
- Quanto maior a renda maior a chance de ser aprovado
- Quanto maior o emprestimo solicitado menos é aprovado
- Todos os inadimplentes não são aprovado
